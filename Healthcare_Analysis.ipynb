```python
from scipy.stats import f_oneway

# One-way ANOVA: Billing by Admission Type
groups = [group['Billing Amount'].values for name, group in df.groupby('Admission Type')]
f_stat, p_val = f_oneway(*groups)
print(f"ANOVA F = {f_stat:.2f}, p = {p_val:.4f}")
```

```python
from scipy.stats import ttest_ind

a = df[df['Admission Type'] == 'Emergency']['Billing Amount']
b = df[df['Admission Type'] == 'Elective']['Billing Amount']

t_stat, p_val = ttest_ind(a, b, equal_var=False)  # Welch’s t-test
print(f"T = {t_stat:.2f}, p = {p_val:.4f}")

```

```python
from scipy.stats import chi2_contingency

contingency = pd.crosstab(df['Gender'], df['Test Results'])
chi2, p, dof, expected = chi2_contingency(contingency)
print(f"Chi-square = {chi2:.2f}, p = {p:.4f}")

```

```python
crosstab = pd.crosstab(df['Gender'], df['Test Results'], normalize='index') * 100
print(crosstab.round(1))
```
